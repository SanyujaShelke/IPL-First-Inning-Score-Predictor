# IPL First Inning Score Predictor

Aim :-
    
To develop a machine learning model to predict the total runs scored by a team in the first innings of an IPL cricket match. Based on various factors such as historical match data, including information about venue, batting and bowling teams, player statistics, and match-specific details.

Dataset - https://raw.githubusercontent.com/SarangGami/IPL-First-Innings-Score-Prediction/main/ipl.csv

Overview :-

This project uses machine learning techniques to predict the outcomes of IPL matches involving 14 teams. The dataset includes IPL match records from 2008 to 2018, and the goal is to accurately predict the outcomes of matches.

Based on the available data from 2008 to 2018, our machine learning model was trained and evaluated on matches involving the 14 teams that were present during that period. We did not include the two new teams introduced in 2022, Gujarat Titans and Lucknow Supergiants, due to the imbalance issue as we did not have sufficient data for these teams.

Therefore, the predictions made by our model should be interpreted with this limitation in mind. Nonetheless, our model can still provide valuable insights into the performance of the 14 teams that were present during the 2008-2018 period and can be used to make informed predictions about future matches involving these teams.

# Data Preparation

```python
# Encoding Categorical Variables
from sklearn.preprocessing import LabelEncoder
```

```python
label_encodings = {}

# Apply Label Encoding and store mappings
for column in data.columns[:5]:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encodings[column] = dict(zip(le.classes_, le.transform(le.classes_)))
```

```python
label_encodings
```

```python
# Splitting Dataset
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('total', axis=1)
y = data['total']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Model Development

# Simple Linear Regression

```python
# Importing Model
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
# Training Model
model1.fit(x_train, y_train)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
y_pred1
```

```python
# checking for scores
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Mean Absolute Error
MAE1 = mean_absolute_error(y_test, y_pred1)
```

```python
# Mean Squared Error
MSE1 = mean_squared_error(y_test, y_pred1)
```

```python
# Root Mean Squared Error
RMSE1 = np.sqrt(MSE1)
```

```python
r2_score1 = r2_score(y_test, y_pred1)
```

```python
print(MAE1)
print(MSE1)
print(RMSE1)
print(r2_score1)
```

```python
model1.predict([[14,5,10,328,201,1,0,0.1,1,0,0,0]])
```

```python
# Let's assume you have received user inputs in variables
venue = 'M Chinnaswamy Stadium'
bat_team = 'Kolkata Knight Riders'
bowl_team = 'Royal Challengers Bangalore'
batsman = 'BB McCullum'
bowler = 'Z Khan'

# Convert user inputs to encoded values using label encodings
venue_encoded = label_encodings['venue'][venue]
bat_team_encoded = label_encodings['bat_team'][bat_team]
bowl_team_encoded = label_encodings['bowl_team'][bowl_team]
batsman_encoded = label_encodings['batsman'][batsman]
bowler_encoded = label_encodings['bowler'][bowler]
print(venue_encoded,bat_team_encoded,bowl_team_encoded,batsman_encoded,bowler_encoded)

# Now you can use these encoded values in your model prediction
# For example, create a new input data point with these encoded values
new_input = [[venue_encoded, bat_team_encoded, bowl_team_encoded, batsman_encoded, bowler_encoded,159,2,16.1,49,1,103,9]]

# Predict using your trained model
predicted_score = model1.predict(new_input)
predicted_score
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor()
model2.fit(x_train, y_train)
y_pred2 = model2.predict(x_test)
MAE2 = mean_absolute_error(y_test, y_pred2)
MSE2 = mean_absolute_error(y_test, y_pred2)
RMSE2 = np.sqrt(MSE2)
r2_score2 = r2_score(y_test, y_pred2)
print(MAE2)
print(MSE2)
print(RMSE2)
print(r2_score2)
```
